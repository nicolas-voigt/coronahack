```python
http_encoding = resp.encoding if 'charset' in resp.headers.get('content-type', '').lower() else None
html_encoding = EncodingDetector.find_declared_encoding(resp.content, is_html=True)
encoding = html_encoding or http_encoding
baseUrl = "https://www.dortmund.de"
locale.setlocale(locale.LC_ALL, 'de_DE')
```

```python
url = 'https://www.dortmund.de/de/leben_in_dortmund/nachrichtenportal/alle_nachrichten/index.jsp?suchbegriff=&stadtbezirk=&thema=&projekt=&tag=&zeitraumVon=21.02.2020&zeitraumBis=22.03.2020&newsSuche=Suche+starten'
resp = requests.get(url)
soup = BeautifulSoup(resp.content, 'lxml', from_encoding=encoding)


dates = []
titles = []
urls = []


for i in soup.findAll("h4",{"class":"headline-medium font-linkred"}):
    templist = list(i.children)
    titles.append(templist[0])

for i in soup.findAll("a",{"class":"col-xs-12 nopadding border-bottom-dotted clearfix col-paddingTop col-paddingBottom linkElement"}):
    urls.append(baseUrl + i.get('href'))

for i in soup.findAll("div",{"class":"col-xs-s-12 col-xs-m-12 col-xs-2 nopadding col-abstandUnten"}):
    for j in i.findAll("div",{"class":"font-grey-3"}):
        strList = j.text.strip().split(".")
        date = datetime(int(strList[2]),int(strList[1]),int(strList[0]))
        dates.append(date)
```

```python
for i in range(0,len(dates)):
    dbInterface.InsertNews("Dortmund","Stadt Dortmund","Stadtregierung",str(dates[i]),str(titles[i]),str(urls[i]),fromFedGovt = False)
```
